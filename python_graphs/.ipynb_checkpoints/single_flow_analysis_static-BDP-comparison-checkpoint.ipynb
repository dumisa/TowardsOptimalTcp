```python
def weighted_average(distribution, weights):
    return round(sum([distribution[i]*weights[i] for i in range(len(distribution))])/sum(weights),2)

def import_csv(csvfilename):
    data = []
    with open(csvfilename, "r", encoding="utf-8", errors="ignore") as scraped:
        reader = csv.reader(scraped, delimiter=' ', skipinitialspace='True')
        next(reader)
        row_index = 0
        for row in reader:
            if row:  # avoid blank lines
                #row_index += 1
                #columns = [str(row_index), row[0], row[1], row[2], row[3]]
                columns = [row[0], row[1]]
                data.append(columns)

    return data

def import_csv2(csvfilename):
    data = []
    with open(csvfilename, "r", encoding="utf-8", errors="ignore") as scraped:
        reader = csv.reader(scraped, delimiter=' ', skipinitialspace='True')
        next(reader)
        row_index = 0
        for row in reader:
            if row:  # avoid blank lines
                #row_index += 1
                #columns = [str(row_index), row[0], row[1], row[2], row[3]]
                columns = [row[0], row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8],
                           row[9], row[10] ]
                data.append(columns)

    return data

def import_csv3(csvfilename):
    data = []
    with open(csvfilename, "r", encoding="utf-8", errors="ignore") as scraped:
        reader = csv.reader(scraped, delimiter=' ', skipinitialspace='True')
        next(reader)
        row_index = 0
        for row in reader:
            if row:  # avoid blank lines
                #row_index += 1
                #columns = [str(row_index), row[0], row[1], row[2], row[3]]
                columns = [row[0], row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8],
                           row[9], row[10], row[11] ]
                data.append(columns)

    return data

def import_csv4(csvfilename):
    data = []
    with open(csvfilename, "r", encoding="utf-8", errors="ignore") as scraped:
        reader = csv.reader(scraped, delimiter=' ', skipinitialspace='True')
        next(reader)
        row_index = 0
        for row in reader:
            if row:  # avoid blank lines
                #row_index += 1
                #columns = [str(row_index), row[0], row[1], row[2], row[3]]
                columns = [row[0], row[1], row[2], row[3], row[4]]
                data.append(columns)

    return data
```

```python
tcp_cca = (['TcpCubic', 'TcpBbr', 'TcpQtCol_1.2minRTT', 'TcpQtCol_1.5minRTT',
            'TcpQtColFair_1.2minRTT', 'TcpQtColFair_1.5minRTT'
           ])
BDP = 172
buffer_size_array = [172, 345, 690, 1036, 1381, 1727, 2072, 2418, 2763, 3109, 3454] 
print(buffer_size_array, tcp_cca)
```

```python
num_flows = 1
cwnd = {}
agg_cwnd_mean = {}
inflight = {}
rtt = {}

for algorithm in tcp_cca:
    cwnd[algorithm] = {}
    agg_cwnd_mean[algorithm] = {}
    inflight[algorithm] = {}
    rtt[algorithm] = {}

    for buffer_size in buffer_size_array:
        cwnd[algorithm][buffer_size] = {}
        agg_cwnd_mean[algorithm][buffer_size] = {}
        inflight[algorithm][buffer_size] = {}
        rtt[algorithm][buffer_size] = {}
        
        agg_cwnd_mean[algorithm][buffer_size] = {'agg_inflight': list(),
                    'aveRtt': list(), 'agg_aveThruput': list(), 'agg_aveGoodput': list(), 
                    'agg_TxPkts': list(), 'PLR': list()}
```

```python
cwnd_files = []
agg_cwnd_files = []
inflight_files = []
rtt_files = []
                                               
for algorithm in tcp_cca:
    print(algorithm)
    cwnd_sim_files = []
    agg_cwnd_files_ = []
    
    for buffer_size in buffer_size_array: 
        agg_cwnd_files_.append("../results/qt_bdp_comparison/" + str(algorithm) + "/" + 
                        str(num_flows)  + "-flows/100Mbps-20ms/" + str(buffer_size) + "p-btlqueue/" + 
                        algorithm + "-" + str(num_flows) + "-flows-100Mbps-20ms-" + str(buffer_size) +
                        "p-" + "Agg" + "-cwnd.data")
    
    agg_cwnd_files.append(agg_cwnd_files_)
```

```python
#print(agg_cwnd_files[0])
#print(agg_cwnd_files[1])
#print(agg_cwnd_files[2])
```

```python
agg_inflight_mean = []
aveRtt_mean = []
agg_aveThruput = []
agg_aveGoodput = []
PLR = []
for algorithm  in tcp_cca:
    index_alg = tcp_cca.index(algorithm)
    metrics_ = []
    
    for buffer_size in buffer_size_array:
        #print(buffer_size_array)
        index_buf = buffer_size_array.index(buffer_size)
        agg_cwnd_data_in = import_csv3(agg_cwnd_files[index_alg][index_buf])

        agg_inflight_ = []
        aveRtt_ = []
        agg_aveThruput_ = []
        agg_aveGoodput_ = []
        agg_TxPkts_ = []
        agg_ReTxPkts_ = []
        PLR_ = []
        
        for row in agg_cwnd_data_in: 
            
            agg_inflight_.append(row[2])
            aveRtt_.append(row[3])
            agg_aveThruput_.append(row[5])
            agg_aveGoodput_.append(row[7])
            agg_TxPkts_.append(row[8])
            agg_ReTxPkts_.append(row[9])
            if float(row[8]) > 0:
                PLR_.append(float(row[9])/float(row[8]))
                        
        agg_inflight_mean.append(agg_inflight_.to_numpy(dtype=float).mean()/1448)
        aveRtt_mean.append(aveRtt_.to_numpy(dtype=float).mean())
        agg_aveThruput.append(agg_aveThruput_.to_numpy(dtype=float)[len(agg_aveThruput)-1])
        agg_aveGoodput.append(agg_aveGoodput_.to_numpy(dtype=float)[len(agg_aveGoodput)-1])
        PLR.append(PLR_.to_numpy(dtype=float)[len(PLR_)-1])
```
